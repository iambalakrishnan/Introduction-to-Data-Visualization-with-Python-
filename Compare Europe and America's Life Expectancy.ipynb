```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Load the datafile
data = pd.read_csv('countries.csv')
```

```python
#preview
data.head()
```

```python
#select 1997 data
data_1997 = data[data['year'] == 1997]
```

```python
data_1997
```

```python
#selecting America's 1997 data
americas_1997 = data_1997[data_1997['continent'] == 'Americas']
```

```python
americas_1997
```

```python
#Selecting Europe 1997 data
europe_1997 = data_1997[data_1997['continent'] == 'Europe']
```

```python
europe_1997
```

```python
#Find the ledngth of America's 1997 data 
print(len(set(americas_1997['country'])))
```

```python
#Find the ledngth of Europe's 1997 data 
print(len(set(europe_1997['country'])))
```

```python
#Finding mean and median for America's 1997 data

print("Mean of life expectancy in America : ", americas_1997['lifeExpectancy'].mean())

print("Median of life expectancy in America : ", americas_1997['lifeExpectancy'].median())
```

```python
#Finding mean and median for Europe's 1997 data

print("Mean of life expectancy in europe : ", europe_1997['lifeExpectancy'].mean())

print("Median of life expectancy in europe : ", europe_1997['lifeExpectancy'].median())
```

```python
#Plotting

plt.subplot(211)
bins = 20
plt.title('Distribution of Life Expectancy')
plt.hist(americas_1997['lifeExpectancy'], bins, range = (55, 85), edgecolor = 'black')
plt.ylabel('Americas')

plt.subplot(212)
plt.hist(europe_1997['lifeExpectancy'], bins, range = (55, 85), edgecolor = 'black')
plt.ylabel('Europe')
plt.show()
```
